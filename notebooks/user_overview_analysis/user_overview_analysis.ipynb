### User Overview analysis 
- Start by identifying the top 10 handsets used by the customers.
- Then, identify the top 3 handset manufacturers
- Next, identify the top 5 handsets per top 3 handset manufacturer
- Make a short interpretation and recommendation to marketing teams


### Import packages

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Load the data

```python
df = pd.read_csv("../../data/tellco_data.csv")
df
```

### the top 10 handsets used by the customers

```python

top_handsets = df['handset_type'].value_counts().head(10)

# Plotting the bar chart
plt.figure(figsize=(10, 6))
top_handsets.plot(kind='bar', color='skyblue')
plt.title('Top 10 Handsets Used by Customers')
plt.xlabel('Handset Type')
plt.ylabel('Number of Users')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

### The top 3 handset manufacturers

```python
df['handset'] = df['handset_manufacturer'] + ' ' + df['handset_type']

top_manufacturers = df['handset_manufacturer'].value_counts().head(3)

# Plotting the bar chart for top manufacturers
plt.figure(figsize=(8, 5))
top_manufacturers.plot(kind='bar', color='orange')
plt.title('Top 3 Handset Manufacturers')
plt.xlabel('Manufacturer')
plt.ylabel('Number of Users')
plt.xticks(rotation=0)  # Keep x-axis labels horizontal
plt.show()
```

### The top 5 handsets per top 3 handset manufacturer


```python
df['handset'] = df['handset_manufacturer'] + ' ' + df['handset_type']

# Identify the top 3 handset manufacturers
top_manufacturers = df['handset_manufacturer'].value_counts().head(3).index

# Create a DataFrame for each top manufacturer
manufacturer_dfs = [df[df['handset_manufacturer'] == manufacturer] for manufacturer in top_manufacturers]

# Plotting the stacked bar chart for each top manufacturer
plt.figure(figsize=(12, 8))

for i, manufacturer_df in enumerate(manufacturer_dfs):
    top_handsets = manufacturer_df['handset_type'].value_counts().head(5)
    plt.bar(top_handsets.index, top_handsets, label=top_manufacturers[i])

plt.title('Top 3 Handset Types for Each Top 3 Handset Manufacturer')
plt.xlabel('Handset Type')
plt.ylabel('Number of Users')
plt.legend()
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

### Interpretation and recommendation

**Interpretation:**

1. **Top 10 Handsets:** Huawei B528S-23A leads the top 10 handsets, followed by popular Apple and Samsung models. The presence of 'undefined' suggests some missing handset types.

2. **Top 3 Handset Manufacturers:** Apple dominates, followed by Samsung and Huawei, indicating a strong preference for Apple devices.

3. **Top 5 Handsets per Manufacturer:** Apple's iPhone models dominate its top 5, while Huawei's B528S-23A stands out, and Samsung's Galaxy S8 is prominent.

**Recommendations:**

1. **Apple Focus:** Emphasize Apple products in marketing, with targeted promotions for newer iPhone models.

2. **Diversification for Samsung and Huawei:** Promote newer models and unique features to diversify handset options.

3. **Customer Engagement:** Engage with customers through surveys and feedback sessions to understand preferences and expectations.
